```python
import json
with open('deps.json') as json_file:
    deps = json.load(json_file)
```

```python
geo_json = {"type": "FeatureCollection", "features": []}
geo_json["features"]

for i in deps:
    
    feature = {"type": "Feature",
               "geometry": {
                   "type": "Polygon",
                   "coordinates": i["coords"][0]},
               "properties": {
                   "title": i["title"],
                   "deputies": ", ".join([i["deputy_name"] for i in i["deputies"]])
               }
              }
    geo_json["features"].append(feature)
```

```python
coordinates_flat = []

for i in geo_json["features"]:
    coordinates = i["geometry"]["coordinates"]
    coordinates_f = []
    for i in coordinates:
        flat = [i["lng"], i["lat"]]
        coordinates_f.append(flat)
    coordinates_f.append(coordinates_f[0])
    coordinates_flat.append(coordinates_f)
    
geo_json
```

```python
for old, new in zip(geo_json["features"], coordinates_flat):
    old["geometry"]["coordinates"] = [new]

with open('geo_json.json', 'w') as outfile:
    json.dump(geo_json, outfile)
```
